# Bayesian Optimization

## Objective Function

The goal is to optimize the critical_temp using Bayesian optimization. The objective function will take the parameters of the superconducting material and return the negative of the critical_temp, as we want to maximize it.

```python
from skopt import BayesSearchCV
from sklearn.ensemble import RandomForestRegressor
```

```python
# Define the model
model = RandomForestRegressor(random_state=random_state)
```

```python
# Define the search space for hyperparameters
search_space = {
    'n_estimators': (50, 500),  # Number of trees in the forest
    'max_depth': (5, 50),        # Maximum depth of the tree
    'min_samples_split': (2, 20), # Minimum number of samples required to split an internal node
    'min_samples_leaf': (1, 20),  # Minimum number of samples required to be at a leaf node
    'max_features': ['sqrt', 'log2', None],  # Number of features to consider when looking for the best split
    'bootstrap': [True, False]  # Whether bootstrap samples are used when building trees
}
```

```python
# Define the Bayesian optimization search
opt = BayesSearchCV(
    model,
    search_space,
    n_iter=50,  # Number of iterations for optimization
    scoring='neg_mean_squared_error',  # Objective function to minimize
    cv=5,  # Cross-validation splitting strategy
    n_jobs=-1,  # Use all available cores
    random_state=random_state
)
```

```python
# Fit the model using Bayesian optimization
opt.fit(X_train, y_train.values.ravel())
```

```python
# Display the best parameters found by Bayesian optimization
print("Best parameters found by Bayesian optimization:")
pprint(opt.best_params_)
```

```python
# Display the best score achieved
print(f"Best score achieved (negative MSE): {opt.best_score_}")
```

```python
# Evaluate the optimized model on the test set
from sklearn.metrics import mean_squared_error
y_pred = opt.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error on test set: {mse}")
```

## Use Optimized Model to Optimize critical_temp using Bayesian Optimization

```python
# Use the optimized model to predict critical_temp
optimized_critical_temp = opt.predict(X_test)
print("Predicted critical_temp using Optimized Model:")
print(optimized_critical_temp)
```

```python
# Visualize the predicted vs actual critical_temp
plt.figure(figsize=(10, 6))
plt.scatter(y_test, optimized_critical_temp, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.title('Predicted vs Actual critical_temp', fontsize=16)
plt.xlabel('Actual critical_temp')
plt.ylabel('Predicted critical_temp')
plt.grid()
plt.show()
```

```python
# First check the actual column names in the DataFrame
print("Actual column names in X_train:")
print(X_train.columns.tolist())
```

```python
# Objective function for Bayesian optimization
def objective_function(params):
    """
    Objective function to minimize the negative of the critical_temp.
    This function takes a the properties and returns the negative critical_temp.
    """
    # Create a DataFrame with the parameters using X_train.columns
    param_dict = {col: [param] for col, param in zip(X_train.columns, params)}
    elemental_properties = pd.DataFrame(param_dict)
    # Predict the critical_temp using the optimized model
    predicted_critical_temp = opt.predict(elemental_properties)
    # Return the negative critical_temp (as we want to maximize it)
    return -predicted_critical_temp[0]
```

```python
# Define the search space for the properties based on the min and max values in the training set
search_space_properties = [(X_train[col].min(), X_train[col].max()) for col in X_train.columns]
```

```python
from skopt import gp_minimize
# Perform Bayesian optimization to find the optimal properties
result = gp_minimize(
    objective_function,
    search_space_properties,
    n_calls=50,  # Number of evaluations
    random_state=random_state,
    verbose=True
)
```

```python
# Neatly display the best parameters and the best predicted critical_temp along with column names
best_params = result.x
best_critical_temp = -result.fun  # Negate the result to get the actual critical_temp
print("Best parameters found by Bayesian optimization:")
for i, param in enumerate(best_params):
    print(f"{X_train.columns[i]}: {param}")
print(f"Best predicted critical_temp: {best_critical_temp}")
```

## Use XGBoost for the Model instead of Random Forest

```python
from xgboost import XGBRegressor
```

```python
# Define the XGBoost model
xgb_model = XGBRegressor(random_state=random_state, n_jobs=-1)
```

```python
# Define the Bayesian optimization search for XGBoost hyperparameters
xgb_search_space = {
    'n_estimators': (50, 500),  # Number of trees in the forest
    'max_depth': (3, 10),        # Maximum depth of the tree
    'learning_rate': (0.01, 0.3, 'uniform'),  # Step size shrinkage used in update to prevent overfitting
    'subsample': (0.5, 1.0, 'uniform'),  # Subsample ratio of the training instances
    'colsample_bytree': (0.5, 1.0, 'uniform'),  # Subsample ratio of columns when constructing each tree
    'gamma': (0, 5),  # Minimum loss reduction required to make a further partition on a leaf
    'reg_alpha': (0, 1),  # L1 regularization term on weights
    'reg_lambda': (0, 1)  # L2 regularization term on weights
}
```

```python
# Define the Bayesian optimization search for XGBoost
xgb_opt = BayesSearchCV(
    xgb_model,
    xgb_search_space,
    n_iter=50,  # Number of iterations for optimization
    scoring='neg_mean_squared_error',  # Objective function to minimize
    cv=5,  # Cross-validation splitting strategy
    n_jobs=-1,  # Use all available cores
    random_state=random_state
)
```

```python
# Fit the XGBoost model using Bayesian optimization
xgb_opt.fit(X_train, y_train.values.ravel())
```

```python
# Display the best parameters found by Bayesian optimization for XGBoost
print("Best parameters found by Bayesian optimization for XGBoost:")
pprint(xgb_opt.best_params_)
```

```python
# Display the best score achieved by XGBoost
print(f"Best score achieved (negative MSE) by XGBoost: {xgb_opt.best_score_}")
```

```python
# Evaluate the optimized XGBoost model on the test set
y_pred_xgb = xgb_opt.predict(X_test)
mse_xgb = mean_squared_error(y_test, y_pred_xgb)
print(f"Mean Squared Error on test set by XGBoost: {mse_xgb}")
```

```python
# Use the optimized XGBoost model to predict critical_temp
optimized_critical_temp_xgb = xgb_opt.predict(X_test)
print("Predicted critical_temp using Optimized XGBoost Model:")
print(optimized_critical_temp_xgb)
```

```python
# Visualize the predicted vs actual critical_temp using XGBoost
plt.figure(figsize=(10, 6))
plt.scatter(y_test, optimized_critical_temp_xgb, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.title('Predicted vs Actual critical_temp using XGBoost', fontsize=16)
plt.xlabel('Actual critical_temp')
plt.ylabel('Predicted critical_temp')
plt.grid()
plt.show()
```

```python
# Objective function for Bayesian optimization with XGBoost
def objective_function_xgb(params):
    """
    Objective function to minimize the negative of the critical_temp using XGBoost.
    This function takes the properties and returns the negative critical_temp.
    """
    # Create a DataFrame with the parameters using X_train.columns
    param_dict = {col: [param] for col, param in zip(X_train.columns, params)}
    elemental_properties = pd.DataFrame(param_dict)
    # Predict the critical_temp using the optimized XGBoost model
    predicted_critical_temp = xgb_opt.predict(elemental_properties)
    # Return the negative critical_temp (as we want to maximize it)
    return -predicted_critical_temp[0]
```

```python
# Define the search space for the properties based on the min and max values in the training set
search_space_properties_xgb = [(X_train[col].min(), X_train[col].max()) for col in X_train.columns]
```

```python
from skopt import gp_minimize
# Perform Bayesian optimization to find the optimal properties using XGBoost
result_xgb = gp_minimize(
    objective_function_xgb,
    search_space_properties_xgb,
    n_calls=50,  # Number of evaluations
    random_state=random_state,
    verbose=True
)
```

```python
# Neatly display the best parameters and the best predicted critical_temp along with column names for XGBoost
best_params_xgb = result_xgb.x
best_critical_temp_xgb = -result_xgb.fun  # Negate the result to get the actual critical_temp
print("Best parameters found by Bayesian optimization for XGBoost:")
for i, param in enumerate(best_params_xgb):
    print(f"{X_train.columns[i]}: {param}")
print(f"Best predicted critical_temp using XGBoost: {best_critical_temp_xgb}")
```
